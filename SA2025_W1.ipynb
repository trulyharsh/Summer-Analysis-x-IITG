### 1. What is name of car that has the highest horsepower?  

```python
df_original = pd.read_csv('Cars.csv')

max_hp_value = df_original['horsepower'].max()

car_with_highest_hp = df_original[df_original['horsepower'] == max_hp_value]
name_of_car = car_with_highest_hp['name'].iloc[0]

print(f"The car with the highest horsepower is: {name_of_car}")
```

### 2. How many cars have mpg ≥ 35?

```python
# Filter the DataFrame for cars with mpg >= 35
cars_mpg_gte_35 = df[df['mpg'] >= 35]

# Count the number of such cars
num_cars_mpg_gte_35 = len(cars_mpg_gte_35)

print(num_cars_mpg_gte_35)
```

### 3. What is the most common origin for cars with horsepower > 100 and weight < 3000?

```python
filtered_cars = df[(df['horsepower'] > 100) & (df['weight'] < 3000)]
most_common_origin = filtered_cars['origin'].mode()[0]
print(most_common_origin)
```

### 4. What is the mean acceleration of cars from Japan? (rounded to 2 decimals)

```python
# Filter cars with origin 'japan'
japan_cars = df[df['origin'] == 'japan']

# Calculate the mean of the 'acceleration' column for Japan cars
mean_acceleration_japan = japan_cars['acceleration'].mean()

# Round the result to two decimal places
rounded_mean_acceleration = round(mean_acceleration_japan, 2)

print(rounded_mean_acceleration)
```

### 5. Which year had the highest average mpg?

```python
# Group by 'model_year' and calculate the mean of 'mpg' for each year
average_mpg_by_year = df.groupby('model_year')['mpg'].mean()

# Find the model_year with the highest average mpg
year_highest_avg_mpg = average_mpg_by_year.idxmax()

print(year_highest_avg_mpg)
```

### Design a multi-line plot using Matplotlib or Seaborn that shows the evolution of average mpg over the years, separately for each origin

```python


# Re-read the CSV file to ensure all original columns are available
df_working = pd.read_csv('Cars.csv')

# Calculate the median mpg
median_mpg = df_working['mpg'].median()

# Filter cars with above-median mpg
above_median_mpg_cars = df_working[df_working['mpg'] > median_mpg].copy()

# Calculate the horsepower to weight ratio for these cars
above_median_mpg_cars['horsepower_to_weight_ratio'] = above_median_mpg_cars['horsepower'] / above_median_mpg_cars['weight']

# Find the maximum ratio among these filtered cars
max_ratio = above_median_mpg_cars['horsepower_to_weight_ratio'].max()

# Find the car(s) with this maximum ratio
best_ratio_cars = above_median_mpg_cars[above_median_mpg_cars['horsepower_to_weight_ratio'] == max_ratio]

# Get the name(s) of the car(s) with the best ratio
car_names = best_ratio_cars['name'].tolist()

print(f"Car(s) with the best horsepower-to-weight ratio and above-median mpg: {car_names}")

# --- New code for multi-line plot ---

# Group by model_year and origin, then calculate the mean mpg
avg_mpg_by_year_origin = df_working.groupby(['model_year', 'origin'])['mpg'].mean().reset_index()

# Create the multi-line plot
plt.figure(figsize=(10, 6)) # Set figure size for better readability
sns.lineplot(data=avg_mpg_by_year_origin, x='model_year', y='mpg', hue='origin', marker='o')

# Add title and labels
plt.title('Average MPG Evolution by Model Year and Origin')
plt.xlabel('Model Year')
plt.ylabel('Average MPG')
plt.xticks(avg_mpg_by_year_origin['model_year'].unique()) # Ensure all years are shown as ticks
plt.grid(True, linestyle='--', alpha=0.7) # Add a grid for better readability
plt.legend(title='Origin') # Add a legend for origins
plt.tight_layout() # Adjust layout to prevent labels from overlapping
plt.show()

```

### We define a “consistent” car model as one that was produced over multiple years and had very low variation in mpg across those years (standard deviation < 1.0).

Tasks:

Identify car names that appear in more than one model_year.

For each such name, compute the standard deviation of mpg across years.

Return the car(s) with the lowest variation in mpg, among those with at least 2 appearances and std(mpg) < 1.0.

Report the model name(s), number of appearances, and the average mpg.

Bonus:
Sort the result by number of appearances (descending), then mpg (descending).

```python


df_analysis = pd.read_csv('Cars.csv')

car_appearances = df_analysis.groupby('name')['model_year'].nunique()
multi_year_cars = car_appearances[car_appearances > 1].index.tolist()

df_multi_year = df_analysis[df_analysis['name'].isin(multi_year_cars)].copy()

consistent_car_data = df_multi_year.groupby('name').agg(
    std_mpg=('mpg', 'std'),
    avg_mpg=('mpg', 'mean'),
    num_appearances=('name', 'count')
).reset_index()

consistent_car_data['std_mpg'] = consistent_car_data['std_mpg'].fillna(0)

consistent_car_data = consistent_car_data[consistent_car_data['std_mpg'] < 1.0]

if not consistent_car_data.empty:
    lowest_variation_std = consistent_car_data['std_mpg'].min()
    lowest_variation_cars = consistent_car_data[consistent_car_data['std_mpg'] == lowest_variation_std]

    final_report = lowest_variation_cars.sort_values(
        by=['num_appearances', 'avg_mpg'],
        ascending=[False, False]
    )

    print("Consistent Car Model(s) with Lowest Variation in MPG:")
    print(final_report[['name', 'num_appearances', 'avg_mpg', 'std_mpg']].to_markdown(index=False, numalign="left", stralign="left"))
else:
    print("No consistent car models found matching the criteria (produced over multiple years with std(mpg) < 1.0).")
```
